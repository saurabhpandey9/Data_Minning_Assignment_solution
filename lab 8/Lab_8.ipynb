```
import numpy as np
import pandas as pd
from numpy import linalg as LA
```

```
df = pd.read_csv("iris data without class label.csv", header = None)
```

```
df.head()
```

```
# covariance matrix
df3 = df.cov()
df3.shape
df3
```

```
# correlation matrix
df.corr()
```

```
# standardization
df2 = (df - df.mean())/df.std()
# df2
```

```
df2
```

```
# covariance matrix
df4 = df2.cov()
```

```
# eigenvector
w, v = LA.eig(df4)
print(w) # eigenvalue
v = v.T
print(v) # eigenvector
```

```
df = np.hstack((v[0].reshape(4, 1), v[1].reshape(4, 1)))
```

```
df.shape
```

```
res = df2.dot(df)
```

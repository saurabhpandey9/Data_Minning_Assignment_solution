```python
import numpy as np
import random
import pandas as pd
import statistics
```

```python
df = pd.read_csv('population.csv')
```

```python
df.head()
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.head()
```

# Inserting Nan Values in the columns of the dataframe

```python
drtA = df
```

```python
#selecting random samples from dataframe  #replace = false means same row is not considered again
drtA['c4'].sample(frac=0.1 , replace = False)   
```

#### Creating Missing Values Dataframe

```python
change1 = drtA.sample(n=20).index  # selecting 10% random values and then replacing them with nan
drtA.loc[change1,['c4']] = np.nan

change2 = drtA.sample(n=20).index
drtA.loc[change2,['child']] = np.nan

change3 = drtA.sample(n=20).index
drtA.loc[change3,['age']] = np.nan

change4 = drtA.sample(n=20).index
drtA.loc[change4,['sal']] = np.nan
```

```python
drtA
```

### Handling Missing Values

```python
dbA,dbB,dbC,dbD,dbE = drtA,drtA,drtA,drtA,drtA
```

```python
#deleted rows with na values
dbA.dropna(inplace = True)
```

```python
# replacing nan with global constants
dbB.fillna("Unknown", inplace = True)
```

```python
# replacing nan with mean
dbC.fillna(dbB.mean(), inplace = True)
```

```python
# replacing nan with median
dbD.fillna(dbB.median(), inplace = True)
```

```python
# replacing by mean of 10 samples of size 20 each

# Handling by replacing missing value with mean of 10 samples (20 elements)

# Sampling or Bootstrapping with replacement and repetition
a_new = df.drop(['c4'],axis='columns')  # 10 samples of size 20
a2 = []
for i in range(10):
    a2.append(a_new.sample(n=20,replace= False))

mn = []
for i in range(10):
    mn.append(a2[i].mean())
print(mn)
```

```python
std1 = statistics.stdev(dbA['sal'])
std2 = statistics.stdev(dbB['sal'])
std3 = statistics.stdev(dbC['sal'])
std4 = statistics.stdev(dbD['sal'])
std5 = statistics.stdev(dbE['sal'])
```
